```python
import pandas as pd
import re
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, ConfusionMatrixDisplay
from scipy.sparse import hstack, csr_matrix
from sklearn.base import BaseEstimator, TransformerMixin
import matplotlib.pyplot as plt
```

```python
# -------- Load and preprocess CEAS_08 dataset (Legitimate + Spam) --------
ceas_df = pd.read_csv('C:/Users/D.Sathiya Pandi/Downloads/CEAS_08.csv')
print(ceas_df.head())
ceas_df = ceas_df.drop_duplicates()
ceas_df = ceas_df.dropna(subset=['body', 'label'])
ceas_df['body_clean'] = ceas_df['body'].astype(str).str.lower().str.strip()

label_mapping_ceas = {'0': 0, 'legitimate': 0, 'ham': 0,
                      '1': 1, 'spam': 1}
ceas_df['label'] = ceas_df['label'].astype(str).str.strip().str.lower().map(label_mapping_ceas)
ceas_df = ceas_df.dropna(subset=['label'])
ceas_df['label'] = ceas_df['label'].astype(int)
```

```python
# -------- Load and preprocess phishing validation dataset (Legitimate + Phishing) --------
phishing_df = pd.read_csv('C:/Users/D.Sathiya Pandi/Downloads/Phishing_validation_emails.csv')
print(phishing_df.head())
phishing_df.rename(columns={'Email Text': 'body_clean'}, inplace=True)
label_mapping_phishing = {'Safe Email': 0, 'Phishing Email': 2}
phishing_df['label'] = phishing_df['Email Type'].map(label_mapping_phishing)
phishing_df = phishing_df.dropna(subset=['label'])
phishing_df['label'] = phishing_df['label'].astype(int)
phishing_df['body_clean'] = phishing_df['body_clean'].astype(str).str.lower().str.strip()
```

```python
# -------- Combine the CEAS_08 and phishing validation datasets --------
combined_df = pd.concat([ceas_df[['body_clean','label']], phishing_df[['body_clean','label']]], ignore_index=True)
combined_df = combined_df.sample(frac=1, random_state=42).reset_index(drop=True)  # shuffle
print(combined_df.head())
```
